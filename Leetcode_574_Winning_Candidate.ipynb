```python
import pandas as pd

def winning_candidate(candidate: pd.DataFrame, vote: pd.DataFrame) -> pd.DataFrame:
   df = pd.merge(vote, candidate, left_on='candidateId', right_on='id', how='left', suffixes=('_v','_c'))
   df2 = df.groupby('name')['id_v'].count().reset_index().sort_values(by='id_v', ascending=False)[['name']].head(1)
   return df2
```
